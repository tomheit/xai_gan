```python
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import colors
import math

import wgan
```

```python
gan100e = wgan.WGan()
genPath = "./wGan_100_epochs/wgan_generator.h5"
discPath = "./wGan_100_epochs/wgan_discriminator.h5"
invPath = "./wGan_100_epochs/wgan_inverter.h5"
gan100e.load_gan(genPath, discPath, invPath)
vector = tf.random.normal(shape = [16, 128])
gen_output = gan100e.generator(vector)
#print(gen_output)
fig, ax = plt.subplots(4,4)
norm = colors.Normalize(vmin = 0, vmax = 1)
for i in range(16):
    img_i = ax[math.floor(i/4), i%4].imshow(gen_output[i,:,:,0], cmap = 'gray', interpolation = 'none')
    img_i.set_norm(norm)
```
